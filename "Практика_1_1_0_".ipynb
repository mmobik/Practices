# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
# Словарь коэффициентов оплаты в зависимости от страны
country = {"Россия": 1.25, "США": 1, "Китай": 1.5}


class Employee:
    # Базовый класс сотрудника
    def __init__(self, hours=None, experience=None, country=None):
        self.hours = hours       # Количество отработанных часов
        self.experience = experience  # Опыт работы в годах
        self.country = country   # Страна проживания


class Manager(Employee):
    # Класс менеджера, наследуется от Employee
    def __init__(self, prepayment, count_of_people, hours, experience, country):
        super().__init__(hours, experience, country)
        self.count_of_people = count_of_people  # Количество подчиненных
        self.prepayment = prepayment  # Размер аванса

    def payment(self):
        # Метод для расчета зарплаты менеджера
        experience = float(self.experience)
        base_salary = self.count_of_people * 1500 + 3000 * self.hours
        if self.prepayment == 0:
            payment = base_salary * (1 + experience / 10) * country[self.country]
        else:
            payment = base_salary * (1 + experience / 10) * country[self.country] - self.prepayment
        print(payment)


class Developer(Employee):
    # Класс разработчика, наследуется от Employee
    def __init__(self, prepayment, hours, experience, country):
        super().__init__(hours, experience, country)
        self.prepayment = prepayment  # Размер аванса

    def payment(self):
        # Метод для расчета зарплаты разработчика
        experience = float(self.experience)
        base_salary = 3000 * self.hours
        if self.prepayment == 0:
            payment = base_salary * (1 + experience / 10) * country[self.country]
        else:
            payment = base_salary * (1 + experience / 10) * country[self.country] - self.prepayment
        print(payment)


# Ввод данных для менеджера
prepayment = int(input("Число аванса или 0 для менеджера: "))
count_of_people = int(input("Количество подчиненных: "))
hours = int(input("Количество отработанных часов: "))
experience = input("Опыт работы (годы): ")
country_input = input(f"Страна проживания (из {list(country.keys())}): ")

if country_input in country:
    person_1 = Manager(prepayment=prepayment, count_of_people=count_of_people, hours=hours, experience=experience,
                       country=country_input)
    person_1.payment()
else:
    print("Неверная страна.")


# Ввод данных для разработчика
prepayment = int(input("Число аванса или 0 для разработчика: "))
hours = int(input("Количество отработанных часов: "))
experience = input("Опыт работы (годы): ")
country_input = input(f"Страна проживания (из {list(country.keys())}): ")

if country_input in country:
    person_2 = Developer(prepayment=prepayment, hours=hours, experience=experience,
                       country=country_input)
    person_2.payment()
else:
    print("Неверная страна.")
```

# Задание 2

**Описание:** Создайте иерархию классов для различных типов транспортных средств (Необходим один родительский класс и 3 дочерних). Реализуйте метод, который позволяет каждому транспортному средству возвращать собственное описание (Метод в каждом классе должен иметь одинаковое название). Продемонстрируйте вызов данного метода для каждого транспортного средства.


Отрабатываемый принцип: Полиморфизм

```python
# Родительский класс Transport
class Transport:
    def __init__(self, speed, type_of_transport, endurance, mobility):
        # Создание атрибутов класса
        self.speed = speed
        self.type_of_transport = type_of_transport
        self.endurance = endurance
        self.mobility = mobility

    def description(self):
        return f"Данные о {self.type_of_transport} еще не сформированы!"


# Определение класса Car, который наследуется от класса Transport
class Car(Transport):
    def __init__(self, wheels, engine, speed, type_of_transport, endurance, mobility):
        # Инициализация атрибутов класса
        self.wheels = wheels
        self.engine = engine
        super().__init__(speed, type_of_transport, endurance, mobility)

    def description(self):
        return (f"Транспорт: {self.type_of_transport}. колес: {self.wheels}, двигатель: {self.engine},"
                f" скорость: {self.speed}, прочность: {self.endurance}, мобильность: {self.mobility}")


# Определение класса Ship, который наследуется от класса Transport
class Ship(Transport):
    def __init__(self, sails, engine, speed, type_of_transport, endurance, mobility):
        # Инициализация атрибутов класса
        self.sails = sails
        self.engine = engine
        super().__init__(speed, type_of_transport, endurance, mobility)

    def description(self):
        return (f"Транспорт: {self.type_of_transport}. парусов: {self.sails}, двигатель: {self.engine},"
                f" скорость: {self.speed}, прочность: {self.endurance}, мобильность: {self.mobility}")


# Определение класса Plane, который наследуется от класса Transport
class Plane(Transport):
    def __init__(self, aircraft_wings, wheels, engine, speed, type_of_transport, endurance, mobility):
        # Инициализация атрибутов класса
        self.wheels = wheels
        self.aircraft_wings = aircraft_wings
        self.engine = engine
        super().__init__(speed, type_of_transport, endurance, mobility)

    def description(self):
        return (f"Транспорт: {self.type_of_transport}. колес: {self.wheels}, крыльев самолета:"
                f" {self.aircraft_wings} двигатель: {self.engine}, скорость: {self.speed}, прочность:"
                f" {self.endurance}, мобильность: {self.mobility}")


# Создание объектов классов
bmw = Car(4, "Турбо", 100, "Машина", 50, 75)

USS_Nimitz_CVN_68 = Ship(0, "2x A4W", 500, "Корабль", 500, 500)

Boeing_747 = Plane(2, 4, "General Electric CF6",
                   1000, "Самолет", 1000, 500)

# Вывод описаний объектов
print(bmw.description())
print(USS_Nimitz_CVN_68.description())
print(Boeing_747.description())
```

# Задание 3

Онлайн-магазин:
- Создайте модель для онлайн-магазина с классами Product, Order, Customer, и ShoppingCart.
- Product включает информацию о цене, наличии на складе и категории товара.
Order обрабатывает процесс покупки, включая расчет цены с учетом скидок и налогов.
- Customer управляет информацией о пользователе и его истории заказов.
- ShoppingCart позволяет добавлять, удалять и обновлять количество товаров перед оформлением заказа.


```python
class Product:  # Класс определяющий параметры товаров и отвечающий за вывод инфы
    def __init__(self, item=None, category=None, price=None, count=None):
        self.item = item  # Название товара
        self.category = category  # Категория товара
        self.price = price  # Цена товара
        self.count = count  # Количество на складе

    def info(self):  # Метод для вывода инфы о товаре
        return f"Продукт: {self.item}, Категория: {self.category}, Цена: {self.price}, Количество: {self.count}"


# Список, состоящий из обьектов класса Product. Представляет список всей техники.
products = [
    Product("Playstation", "Приставки", 65000, 10),
    Product("Iphone 13", "Смартфоны", 80000, 20),
    Product("Macbook 2020", "Ноутбуки", 120000, 5),
    Product("Ipad air 2020", "Телевизоры", 90000, 15),
    Product("Samsung watch 5", "Смарт-часы", 20000, 10)
]


class Order:  # Класс оплаты
    def __init__(self, cart, customer):
        self.cart = cart  # Корзина покупок
        self.customer = customer  # Объект клиента

    def buying_process(self):  # Метод для оформления покупки
        print("Добро пожаловать в раздел оплаты!")
        print("Для оформления заказа, пожалуйста, выберите товар из корзины.")
        print("Список товаров из корзины: ", end="")
        for product in self.cart.product_in_cart:
            print(f"{product.item} ", end="")
        if not self.cart.product_in_cart:
            print("Корзина пуста.\n")
        else:
            product_selection = input("Выберете продукт: ")
            count = 0
            for product in self.cart.product_in_cart:
                if product.item == product_selection:
                    count += 1
                    print("Инициализируем процесс оплаты...")
                    print(product.info())
                    print("Скидки - Нет")
                    print("Налог - 5%")
                    total_price = product.price * 1.05
                    print(f"Итоговая сумма: {total_price}")
                    choice = input("Для оплаты нажмите любую кнопку для отмены напишите stop").lower()
                    if choice == "stop":
                        break
                    else:
                        for pd in products:
                            if pd.item == product.item:  # Изменено на pd.item
                                if pd.count > 0:
                                    pd.count -= 1
                                    self.customer.add_order(product.item, total_price)  # Добавляем заказ в историю
                                else:
                                    print("Товар закончился!")
                                    break
                        print(f"Вы приобрели: {product.item}")
                        break
            if count == 0:
                print("Выбран неверный продукт.\n")


class ShoppingCart:  # Класс для управления корзиной пользователя
    def __init__(self):
        self.product_in_cart = []  # Поле, которое дает обьекту характеристику добавленных продуктов

    @staticmethod
    def shoppingcart_options():  # Метод для отображения опций
        print("Добро пожаловать в корзину покупок!")
        print("Список доступных функций:")
        print("1.Посмотреть доступные товары.")
        print("2.Добавить товар в корзину.")
        print("3.Удалить товар из корзины.")
        print("4.Выйти из корзины товаров.\n")

    def available_products(self):  # Метод для выбора опции
        ShoppingCart.shoppingcart_options()
        while True:
            try:
                choice = int(input("Введите опцию: "))
                if choice == 1:
                    self.show_products()  # Показать товары
                elif choice == 2:
                    self.add_product()  # Добавить товары
                elif choice == 3:
                    self.remove_product()  # Удалить товары
                elif choice == 4:
                    print("")
                    break  # Выйти из функции
            except ValueError:
                print("Выберете номер опции!\n")

    @staticmethod
    def show_products():  # Метод показывает все товары магазина
        for product in products:  # Перебираем все продукты магазина из списка и выводим их
            print(product.info())
        print()

    def add_product(self):  # Метод для добавления товара в корзину
        adding_product = input("Введите полное название продукта: ")
        count = 0
        for product in products:
            if product.item == adding_product:
                count += 1
                self.product_in_cart.append(Product(product.item, product.category, product.price, 1))
                print("Продукт успешно добавлен в корзину!\n")
                break
        if count == 0:
            print("Продукт не найден.\n")

    def remove_product(self):
        removing_product = input("Введите полное название продукта: ")
        count = 0
        for product in self.product_in_cart:
            if removing_product == product.item:
                count += 1
                self.product_in_cart.remove(product)
                print("Продукт удален из корзины!\n")
                break
        if count == 0:
            print("Продукта нету в корзине.\n")


class Customer:
    def __init__(self):
        self.order_history = []  # Список для хранения истории заказов

    def add_order(self, product_name, total_price):
        self.order_history.append({"product": product_name, "price": total_price})  # Добавляем заказ

    def show_history(self):
        print("История заказов:")
        for order in self.order_history:
            print(f"Продукт: {order['product']}, Сумма: {order['price']:.2f}руб")


# Основной код
User_1 = ShoppingCart()
User_1.available_products()
customer = Customer()  # Создаем объект клиента
User_2 = Order(User_1, customer)  # Передаем корзину и клиента в класс Order
User_2.buying_process()

# Показываем историю заказов после завершения процесса покупки
customer.show_history()
```

# Задание 4

Симулятор космического корабля:
- Создайте симулятор управления космическим кораблем с классами SpaceShip, CrewMember, и Mission.
- SpaceShip имеет атрибуты для управления топливом, состоянием корпуса, и текущей скоростью.
- CrewMember контролирует здоровье, навыки, и роли в команде (например, пилот, инженер).
- Mission определяет цели, ресурсы, и возможные события (например, аварии, встречи с астероидами).

```python
import random


class CrewMember:
    def __init__(self, name, health, skills, role):
        self.name = name  # Имя члена экипажа
        self.health = health  # Здоровье
        self.skills = skills  # Навыки
        self.role = role  # Роль в команде

    def heal(self, amount):
        if amount > 0:  # Лечение может быть только положительным
            self.health = min(self.health + amount, 100)  # Лечение до 100
            print(f"{self.name} был вылечен на {amount} единиц. Текущее здоровье: {self.health}.")
        else:
            print("Вы не можете отравить члена экипажа!")

    def status(self):
        print(f"Имя: {self.name}, Здоровье: {self.health}, Роль: {self.role}, Навыки: {', '.join(self.skills)}")


class SpaceShip:
    def __init__(self, fuel, hull_strength, speed):
        self.fuel = fuel  # Топливо
        self.hull_strength = hull_strength  # Состояние корпуса
        self.speed = speed  # Скорость

    def fly(self, distance):
        fuel_needed = distance * self.speed * 0.1  # Топливо для полета
        if self.fuel >= fuel_needed:
            self.fuel -= fuel_needed
            print(f"Корабль успешно пролетел {distance} км.")
            self.random_event()  # Случайное событие после полета
        else:
            print("Недостаточно топлива для полета!")

    def repair(self, amount):
        if amount > 0:  # Проверка на положительное значение
            repair_amount = min(amount, 100 - self.hull_strength)  # Ремонт до 100
            self.hull_strength += repair_amount
            print(f"Корпус корабля отремонтирован на {repair_amount} единиц. Текущее состояние: {self.hull_strength}.")

    def random_event(self):
        events = ["Встреча с астероидом", "Нападение пришельцев", "Обнаружение редкого ресурса"]
        event = random.choice(events)
        print(f"Событие: {event}")

        if event == "Встреча с астероидом":
            damage = random.randint(5, 20)
            self.hull_strength -= damage
            print(f"Корабль получил повреждение в размере {damage} единиц.")
        elif event == "Нападение пришельцев":
            damage = random.randint(10, 30)
            self.hull_strength -= damage
            print(f"Пришельцы атаковали! Корабль получил {damage} единиц повреждений.")
            attack_crew(damage)  # Уменьшаем здоровье членов экипажа
        elif event == "Обнаружение редкого ресурса":
            found_resources = random.randint(1, 5)
            print(f"Вы обнаружили {found_resources} единиц редкого ресурса!")

    def status(self):
        print(f"Топливо: {self.fuel}, Состояние корпуса: {self.hull_strength}, Скорость: {self.speed}")


def attack_crew(damage):
    for member in crew:
        member.health -= damage // len(crew)  # Делим урон между членами экипажа
        print(f"{member.name} получил урон. Текущее здоровье: {member.health}.")
        if member.health <= 0:  # Проверка на уничтожение
            print(f"{member.name} был уничтожен!")


class Mission:
    def __init__(self, objectives):
        self.objectives = objectives  # Цели миссии
        self.completed_objectives = 0  # Число выполненных миссий

    def complete_objective(self):
        if self.completed_objectives < len(self.objectives):
            print(f"Цель '{self.objectives[self.completed_objectives]}' выполнена!")
            self.completed_objectives += 1  # Увеличиваем количество выполненных целей

    def is_completed(self):
        return self.completed_objectives == len(self.objectives)  # Проверка завершения миссии


crew = [  # Список членов экипажа
    CrewMember(name="Алекс", health=100, skills=["пилотирование", "навигация"], role="пилот"),
    CrewMember(name="Ирина", health=100, skills=["ремонт", "инженерия"], role="инженер"),
    CrewMember(name="Максим", health=100, skills=["боевые искусства", "тактика"], role="защитник"),
    CrewMember(name="Елена", health=100, skills=["медицина", "психология"], role="врач"),
    CrewMember(name="Сергей", health=100, skills=["компьютерные науки", "программирование"], role="техник"),
]


def main():
    spaceship = SpaceShip(fuel=1000, hull_strength=100, speed=10)  # Создаем космический корабль

    missions = [  # Список миссий
        Mission(objectives=["Исследовать новую планету", "Собрать образцы", "Вернуться домой"]),
        Mission(objectives=["Провести исследование астероида", "Собрать образцы", "Вернуться на базу"]),
        Mission(objectives=["Найти редкий ресурс", "Доставить его на базу"]),
    ]

    current_mission = None  # Текущая миссия

    while True:
        print("\nВыберите действие:")
        print("1. Проверить статус корабля")
        print("2. Проверить статус членов экипажа")
        print("3. Полететь")
        print("4. Ремонтировать корабль")
        print("5. Начать новую миссию")
        print("6. Завершить цель миссии")
        print("7. Выйти из игры")

        choice = input("Введите номер действия: ")

        if choice == "1":
            spaceship.status()

        elif choice == "2":
            for member in crew:
                member.status()
            heal_choice = input("Хотите вылечить члена экипажа? (да/нет): ")
            if heal_choice.lower() == "да":
                member_name = input("Введите имя члена экипажа: ")
                heal_amount = int(input("Введите количество здоровья для восстановления: "))
                for member in crew:
                    if member.name.lower() == member_name.lower():
                        member.heal(heal_amount)
                        break
                else:
                    print("Член экипажа с таким именем не найден.")

        elif choice == "3":
            distance = int(input("Введите расстояние для полета (в км): "))
            spaceship.fly(distance)
            if current_mission:  # Проверяем, есть ли текущая миссия
                if random.random() < 0.5:  # 50% шанс встретить планету
                    print("Вы встретили планету! Высаживаемся для выполнения миссии.")
                    current_mission.complete_objective()
                    if current_mission.is_completed():
                        print("Миссия завершена!")
                        current_mission = None

        elif choice == "4":
            amount = int(input("Введите количество ремонта: "))
            spaceship.repair(amount)

        elif choice == "5":
            if current_mission is None:
                current_mission = random.choice(missions)  # Начинаем новую миссию
                print(f"Начата новая миссия: {', '.join(current_mission.objectives)}")
            else:
                print("У вас уже есть активная миссия.")

        elif choice == "6":
            if current_mission and current_mission.is_completed():
                print("Вы завершили миссию!")
                current_mission = None
            else:
                print("Невозможно завершить миссию. Убедитесь, что все цели выполнены.")

        elif choice == "7":
            print("Выход из игры...")
            break
        else:
            print("Неверный выбор. Пожалуйста, попробуйте снова.")


main()  # Запуск основной функции
```
