# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random

# Информация о герое
hero_info = {
    "name": "",
    "armor": "",
    "weapon": "",
    "health": 0,
    "damage_range": (8, 15),  # Диапазон урона героя
    "strong_attack_range": (15, 30),  # Диапазон урона сильной атаки
    "potions": {"лечение": 2, "сила": 1},
    "strong_attack_used": False  # Флаг, указывающий, использовалась ли сильная атака
}

# Инвентарь
inventory = {
    "доспехи": {
        "кожаный сет": {"уровень_защиты": 10, "прочность": 100, "доп_здоровье": 5},
        "медный сет": {"уровень_защиты": 15, "прочность": 120, "доп_здоровье": 10},
        "стальной сет": {"уровень_защиты": 25, "прочность": 150, "доп_здоровье": 15}
    },
    "оружие": {
        "меч": {"урон": 15, "дальность": 1, "прочность": 100},
        "лук": {"урон": 10, "дальность": 15, "прочность": 80}
    }
}

# Информация о драконе
dragon_info = {
    "name": "Дракон",
    "health": 150,
    "damage_range": (10, 20),  # Диапазон урона дракона
    "fire_breath_damage": (20, 30),  # Урон от огненного дыхания
    "defense": 5,  # Защита дракона
    "attacks": ["Удар хвостом", "Огненное дыхание"]
}


def create_hero(name):
    """Создает героя с заданным именем и случайным здоровьем."""
    hero_info["name"] = name
    hero_info["health"] = random.randint(50, 100)
    print(f"Вы создали героя с именем {name} и {hero_info['health']} здоровья.")


def equip_armor_and_weapon():
    """Позволяет игроку выбрать броню и оружие из инвентаря."""
    while True:
        print("Доступная броня:")
        for armor in inventory["доспехи"]:
            print(armor)
        selected_armor = input("Выберите доступную броню: ").lower()
        if selected_armor in inventory["доспехи"]:
            hero_info["armor"] = selected_armor
            hero_info["health"] += inventory["доспехи"][selected_armor]["доп_здоровье"]
            print(f"Вы надели {selected_armor}. Общее количество здоровья: {hero_info['health']}\n")
            break
        else:
            print("Неверный выбор. Пожалуйста, выберите броню снова.\n")

    while True:
        print("Доступное оружие:")
        for weapon in inventory["оружие"]:
            print(weapon)
        selected_weapon = input("Выберите доступное оружие: ").lower()
        if selected_weapon in inventory["оружие"]:
            hero_info["weapon"] = selected_weapon
            print(f"Вы взяли {selected_weapon}\n")
            break
        else:
            print("Неверный выбор. Пожалуйста, выберите оружие снова.\n")


def open_inventory():
    """Открывает инвентарь игрока для использования зелий."""
    print("\nВаш инвентарь:")
    for potion, count in hero_info["potions"].items():
        if count > 0:
            print(f"{potion} (количество: {count})")
    potion_choice = input("Выберите зелье для использования: ").lower()
    use_potion(potion_choice)


def use_potion(potion_choice):
    """Использует выбранное зелье, если оно доступно."""
    if potion_choice in hero_info["potions"] and hero_info["potions"][potion_choice] > 0:
        if potion_choice == "лечение":
            hero_info["health"] += 20
            print(f"{hero_info['name']} использует зелье лечения. Текущее здоровье: {hero_info['health']}.")
        elif potion_choice == "сила":
            hero_info["damage_range"] = (hero_info["damage_range"][0] + 5, hero_info["damage_range"][1] + 5)
            print(f"{hero_info['name']} использует зелье силы. Текущий диапазон урона: {hero_info['damage_range']}.")
        hero_info["potions"][potion_choice] -= 1
    else:
        print("Неверный выбор или недостаточно зелий.")


def battle():
    """Основной цикл битвы между героем и драконом."""
    print(f"Начинается битва между {hero_info['name']} и {dragon_info['name']}!")
    while hero_info["health"] > 0 and dragon_info["health"] > 0:
        print(f"{hero_info['name']} здоровье:"
              f" {hero_info['health']} против {dragon_info['name']} здоровье: {dragon_info['health']}\n")
        action = input("1: Обычная атака / 2: Сильная атака / 3: Использовать зелье\nВыберите действие: ").lower()

        if action == "1":
            damage_dealt = random.randint(*hero_info["damage_range"])
            print(f"\n{hero_info['name']} наносит {damage_dealt} урона {dragon_info['name']}.")
            dragon_info["health"] -= damage_dealt

            # Уменьшаем прочность оружия
            weapon_durability = inventory["оружие"][hero_info["weapon"]]["прочность"]
            inventory["оружие"][hero_info["weapon"]]["прочность"] -= 1
            print(f"Прочность {hero_info['weapon']}:"
                  f" {inventory['оружие'][hero_info['weapon']]['прочность']}/{weapon_durability}")

            # Проверка на износ оружия
            if inventory["оружие"][hero_info["weapon"]]["прочность"] <= 0:
                print(f"{hero_info['name']} потерял {hero_info['weapon']} из-за износа. Выберите новое оружие!")
                hero_info["weapon"] = ""

            hero_info["strong_attack_used"] = False  # Сброс флага после обычной атаки

        elif action == "2":
            if hero_info["strong_attack_used"]:
                print("Сильная атака уже использована. Пожалуйста, выберите обычную атаку или зелье.")
                continue
            strong_damage_dealt = random.randint(*hero_info["strong_attack_range"])
            print(f"\n{hero_info['name']} использует сильную атаку и наносит"
                  f" {strong_damage_dealt} урона {dragon_info['name']}.")
            dragon_info["health"] -= strong_damage_dealt

            # Уменьшаем прочность оружия
            weapon_durability = inventory["оружие"][hero_info["weapon"]]["прочность"]
            inventory["оружие"][hero_info["weapon"]]["прочность"] -= 2  # Сильная атака уменьшает прочность больше
            print(f"Прочность {hero_info['weapon']}:"
                  f"{inventory['оружие'][hero_info['weapon']]['прочность']}/{weapon_durability}")

            # Проверка на износ оружия
            if inventory["оружие"][hero_info["weapon"]]["прочность"] <= 0:
                print(f"{hero_info['name']} потерял {hero_info['weapon']} из-за износа. Выберите новое оружие!")
                hero_info["weapon"] = ""

            hero_info["strong_attack_used"] = True  # Сохранение флага о том, что сильная атака использована

        elif action == "3":
            open_inventory()
            continue  # Возвращаемся к выбору действия после использования зелья

        if dragon_info["health"] <= 0:
            print(f"{dragon_info['name']} повержен! {hero_info['name']} победил!")
            break

        # Дракон атакует
        dragon_damage = random.randint(*dragon_info["damage_range"])
        print(f"{dragon_info['name']} наносит {dragon_damage} урона {hero_info['name']}.")
        hero_info["health"] -= dragon_damage

        # Уменьшаем прочность брони
        if hero_info["armor"] is not "":
            armor_durability = inventory["доспехи"][hero_info["armor"]]["прочность"]
            inventory["доспехи"][hero_info["armor"]]["прочность"] -= 1
            print(f"Прочность {hero_info['armor']}:"
                  f" {inventory['доспехи'][hero_info['armor']]['прочность']}/{armor_durability}")

            # Проверка на износ брони
            if inventory["доспехи"][hero_info["armor"]]["прочность"] <= 0:
                print(f"{hero_info['name']} потерял {hero_info['armor']} из-за износа. Выберите новую броню!")
                hero_info["armor"] = ""

        if hero_info["health"] <= 0:
            print(f"{hero_info['name']} повержен! {dragon_info['name']} победил!")
            break


def history():
    """Выводит мини-сюжет игры."""
    print("\nМини-сюжет:")
    print(
        f"В темном лесу, окруженном горными вершинами, жил герой по имени {hero_info['name']}. \n"
        f"Слухи о страшном драконе, который терроризировал окрестные деревни, достигли его ушей. \n"
        "Каждый вечер дракон появлялся на небе, его огненное дыхание разрывалось над землей, и страх охватывал людей.\n"
        "Однажды, когда герой сидел у костра, он услышал рассказ о том, как дракон похитил принцессу из королевства. \n"
        "Гнев и решимость взбурлили в его душе. Он знал, что должен бросить вызов этому чудовищу.\n"
        "Собрав свою броню и оружие, он поклялся, что освободит принцессу и покончит с правлением страха. \n"
        "Его путь был полон опасностей, но он был готов сразиться с драконом, чтобы вернуть мир в свои земли.\n"
    )


# Основной игровой процесс
history()
create_hero(input("Введите имя героя: "))
equip_armor_and_weapon()
battle()
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests

pokemon_data = [{"pokemons": []}]  # Словарь для хранения данных покемонов


def add_pokemon(name):  # Функция добавления покемона
    if name not in pokemon_data[0]["pokemons"]:  # Проверка на наличие
        response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{name}')
        if response.status_code == 200:  # Проверка успешности
            response = response.json()
            # Добавляем имя покемона
            pokemon_data[0]["pokemons"].append(response['name'])
            print("\nПокемон добавлен в команду!")
            print(f"Список ваших покемонов: "
                  f"{pokemon_data[0]['pokemons']}\n")
        else:
            print("\nПокемон не найден. "
                  "Проверьте имя.\n")
    else:
        print("\nУ вас уже есть этот покемон "
              "в команде.\n")


def remove_pokemon(name):  # Функция удаления покемона
    if name in pokemon_data[0]["pokemons"]:  # Проверка на наличие
        pokemon_data[0]["pokemons"].remove(name)
        print(f"\nПокемон {name} успешно "
              "исключен из команды.")
        print(f"Список ваших покемонов: "
              f"{pokemon_data[0]['pokemons']}\n")
    else:
        print("\nПокемон не найден. "
              "Проверьте имя.\n")


def check_pokemons_stats():  # Функция проверки статистики
    if not pokemon_data[0]["pokemons"]:  # Проверка на пустоту
        print("Ваша команда пуста.\n")
        return

    for pokemons_name in pokemon_data[0]["pokemons"]:
        response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemons_name}')
        if response.status_code == 200:  # Проверка успешности
            pokemon = response.json()
            types = [style['type']['name'] for style in pokemon['types']]
            print(f"\nИмя: {pokemon['name'].capitalize()}, "
                  f"Типы: {', '.join(types)}")
            print(f"HP: {pokemon['stats'][5]['base_stat']}, "
                  f"Атака: {pokemon['stats'][4]['base_stat']}")
            print(f"Защита: {pokemon['stats'][3]['base_stat']}, "
                  f"Специальная Атака: {pokemon['stats'][1]['base_stat']}")
            print(f"Специальная Защита: {pokemon['stats'][2]['base_stat']}, "
                  f"Скорость: {pokemon['stats'][0]['base_stat']}\n")
        else:
            print(f"\nНе удалось получить данные для "
                  f"покемонов {pokemons_name}.\n")


def check_pokemon(name):  # Функция поиска покемона
    response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{name}')
    if response.status_code == 200:  # Проверка успешности
        pokemon = response.json()
        types = [style['type']['name'] for style in pokemon['types']]
        print(f"\nИмя: {pokemon['name'].capitalize()}, "
              f"Типы: {', '.join(types)}")
        print(f"HP: {pokemon['stats'][5]['base_stat']}, "
              f"Атака: {pokemon['stats'][4]['base_stat']}")
        print(f"Защита: {pokemon['stats'][3]['base_stat']}, "
              f"Специальная Атака: {pokemon['stats'][1]['base_stat']}")
        print(f"Специальная Защита: {pokemon['stats'][2]['base_stat']}, "
              f"Скорость: {pokemon['stats'][0]['base_stat']}\n")
    else:
        print(f"\nНе удалось получить данные для "
              f"покемона {name}.\n")


def training_battle(name_1, name_2):  # Функция для боя
    if name_1 not in pokemon_data[0]["pokemons"] or \
       name_2 not in pokemon_data[0]["pokemons"]:
        print("Один или оба покемона не найдены "
              "в вашей команде. Проверьте имена.\n")
        return

    response_1 = requests.get(f'https://pokeapi.co/api/v2/pokemon/{name_1}')
    response_2 = requests.get(f'https://pokeapi.co/api/v2/pokemon/{name_2}')

    if response_1.status_code == 200 and \
       response_2.status_code == 200:  # Проверка успешности
        pokemon_1 = response_1.json()
        pokemon_2 = response_2.json()

        hp_1 = pokemon_1['stats'][5]['base_stat']
        attack_1 = pokemon_1['stats'][4]['base_stat']
        defense_1 = pokemon_1['stats'][3]['base_stat']

        hp_2 = pokemon_2['stats'][5]['base_stat']
        attack_2 = pokemon_2['stats'][4]['base_stat']
        defense_2 = pokemon_2['stats'][3]['base_stat']

        score_1 = (attack_1 + defense_1) / 2  # Сила 1-го покемона
        score_2 = (attack_2 + defense_2) / 2  # Сила 2-го покемона

        print(f"\nБой между {pokemon_1['name'].capitalize()} и "
              f"{pokemon_2['name'].capitalize()}!")

        if score_1 > score_2:
            print(f"{pokemon_1['name'].capitalize()} побеждает!\n")
        elif score_1 < score_2:
            print(f"{pokemon_2['name'].capitalize()} побеждает!\n")
        else:
            print("Ничья!\n")

    else:
        print("\nНе удалось получить данные для "
              "одного или обоих покемонов.\n")


def main():  # Основная функция
    print("Добро пожаловать в наше приложение!\n")
    while True:
        print("Список доступных команд:")
        print("1. Добавить покемона в команду.")
        print("2. Удалить покемона из команды.")
        print("3. Проверить статистику покемонов.")
        print("4. Найти покемона по имени.")
        print("5. Устроить тренировочный бой")
        print("6. Выйти.\n")
        choice = input("Выберите действие: ")

        if choice == "1":
            add_pokemon(name=input("Введите имя покемона: ").lower())
        elif choice == "2":
            remove_pokemon(name=input("Введите имя покемона: ").lower())
        elif choice == "3":
            check_pokemons_stats()
        elif choice == "4":
            check_pokemon(name=input("Введите имя покемона: ").lower())
        elif choice == "5":
            training_battle(name_1=input("Введите имя 1 покемона: ").lower(),
                            name_2=input("Введите имя 2 покемона: ").lower())
        elif choice == "6":
            print("Спасибо за использование приложения!")
            break
        else:
            print("\nДанной команды нет, повторите "
                  "свой запрос.\n")


main()  # Запуск основной функции
````
